```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/zakkiya/dataminingProject/master/drug200%20(1).csv')
df
```

```python
df.head()
```

```python
Sex = pd.get_dummies(df.Sex)
df = df.drop(columns = ['Sex'])
df[["F", "M"]] = Sex
```

```python
df
```

```python
BP = pd.get_dummies(df.BP)
df = df.drop(columns = ['BP'])
df[["HIGH_BP", "LOW_BP", "NORMAL_BP"]] = BP
```

```python
df
```

```python
Cholesterol = pd.get_dummies(df.Cholesterol)
df = df.drop(columns = ['Cholesterol'])
df[["HIGH_Cl", "NORMAL_CL"]] = Cholesterol
```

```python
df
```

```python
X = df.drop(columns = ['Drug', 'Age','Na_to_K'])
y = df.Drug
X
```

```python
df.groupby('Drug')['Drug'].count()
```

```python
L = df[["Age","Na_to_K"]] 
L
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled= scaler.fit_transform(L)
features_names = L.columns.copy()
scaled_features = pd.DataFrame(scaled, columns=features_names)
scaled_features.head(10)
```

```python
scaler = pd.concat([scaled_features,X], axis=1)
scaler
```

```python
import joblib
filename = "drugsave"
joblib.dump(scaler, filename)
```

```python
cls = df['Drug'].values
```

```python
from sklearn import preprocessing
y = preprocessing.LabelEncoder()
cls_encoder = y.fit_transform(cls)
cls_encoder
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(scaler, cls_encoder, test_size=0.3, random_state=1)
```

knn

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report

# my_param_grid = {'n_neighbors':[2,3,5,7], 'weights': ['distance', 'uniform']}
# GridSearchCV(estimator=KNeighborsClassifier(), param_grid=my_param_grid, refit=True, verbose=3, cv=3)
# knn = GridSearchCV(KNeighborsClassifier(), param_grid=my_param_grid, refit=True, verbose=3, cv=3)

knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
```

```python
#save model knn
filenameModelKnnNorm = 'modelKnnNorm.pkl'
joblib.dump(knn, filenameModelKnnNorm)
```

```python
print("Score data latih = ",knn.score(X_train, y_train))
print("Score data test = ",knn.score(X_test,y_test))
akurasi = knn.score(X_test, y_test) * 100
print("Hasil akurasi = ", round(akurasi,2), "%")
prediksi_knn = knn.predict([[0.135593,0.596848,	1,	0,	1,	0,	0,	1,	0]])
print("Hasil prediksi = ", prediksi_knn[0])
```
